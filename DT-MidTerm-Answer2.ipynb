#### 6. 문자열 안에 있는 문자의 개수, 숫자의 개수, 공백의 개수를 계산하는 프로그램을 작성하여 보자.

<pre>

</pre>

```python
## 이 부분은 건드리지 말것
sentence=input("문자열을 입력하시오: ")
table={"alphas":0, "digits":0, "spaces":0}

## 풀이
for i in sentence:
    if i.isalpha():
        table["alphas"]+=1
    if i.isdigit():
        table["digits"]+=1
    if i.isspace():
        table["spaces"]+=1

print(table)
    
```

본인이 생각한 아이디어보다 전화번호부를 만들고, 연락처 검색 함수를 만드는게 더 재미있을 것 같다는 생각이 들었습니다. 정말 재미있는 문제인지 한번 풀어보도록 합시다

##### 연락처
- 윤승식 : 010-3348-2859
- 허준 : 010-7708-5617
- 신민철 : 010-4729-8241

> ###### 요구사항
> 1. 위에 제시된 연락처만 이용하여 적절한 자료형을 이용하여 만들어주세요.(이름 중복 등은 고려하지 않도록 합니다)
> 2. 연락처에 기록된 전화번호를 검색하는 contact_search() 함수를 만들어주세요. 함수의 결과물로 전화번호를 **리턴** 할 수 있도록 해주세요
> 3. 해당 함수를 통해 연락처를 검색할 때, 이름의 일부만 입력해도 검색이 가능하도록 해 주세요. 

>    *예시: 준 -> 010-7708-5617, 신민철 -> 010-4729-8241, 윤 -> 010-3348-2859*


```python
phone = {"윤승식":"010-3348-2859", "허준":"010-7708-5617", "신민철":"010-4729-8241"}
def contact_search(txt, phone):
    ans = []
    for key, value in phone.items():
        if txt in key:
            ans.append(value)
    return ans
```

```python
print(contact_search("준", phone))
print(contact_search("신민철", phone))
print(contact_search("윤", phone))
```

스키테일 암호를 만드는 방법은 간단합니다. 메시지의 간격(n)을 설정하고, 그 사이에 의미없는 문자열을 삽입하면 되는 것이죠.

윤승식 연구원은 여자친구와의 비밀 메시지를 위해 메시지의 간격을 설정하고, 스키테일 암호를 이용하기로 결정하였습니다.

자, 이제 윤승식 연구원을 위해 간격을 정해주면 스키테일 암호문을 만들어내는 함수 encrypt()와, 이런 암호문을 해독하는 함수 decrypt()를 작성해 봅시다

> ###### 요구사항
> 1. 문자열과, 막대기의 굵기를 입력받아 스키테일 암호문을 암호화하는 함수 encrypt(string, n)과, 스키테일 암호문을 풀어내는 함수 decrypt(string, n)을 작성하세요.
> 2. 사이에 집어넣을 임의의 문자열은 rand_str 함수를 이용하여 생성해주세요.
> 3. 함수의 결과물로 string을 return해주세요.

```python
# 이부분은 건드리지 말것, 한번 실행하고 문제 풀기
import random

character = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
def rand_str(length, characters=character):
    """
    rand_str(length, characters)
    characters에 지정되어 있는 문자열을 임의로 추출하여 length 길이만큼 추출해줍니다.
    characters를 지정하지 않을 시 기본값으로 영어 알파벳 대,소문자에서 임의의 문자를 추출합니다.
    """
    #return "".join([random.choice(characters) for _ in range(length)])
    return "".join(random.choices(characters, k=length))
```

```python
rand_str(30) 
```

```python
def encrypt(string, n):
    message=""
    for i in range(0,len(string)):
        message+=string[i]
        message+=rand_str(n)
    return message

def decrypt(string, n):
    message=""
    for i in range(0,len(string), n+1):
        message+=string[i]
    return message
```

```python
enc = encrypt("ManchesterUnited", 5)
dec = decrypt(enc, 5)

print(enc)
print(dec)
```

5행시를 넣으면(적절한 구조를 생각해서 파라미터로 넣으세요) 첫글자를 확인해서 "정건용바보"가 맞는지 확인하는 check5() 함수를 만들어주세요. 정건용바보가 맞다면 True, 아니면 False를 반환하도록 합니다.

**위에 있는 시 말고, 다른 시를 만들어서 검사할 수 있도록 합니다. 우리 건용이는 소중하니까요!**

```python
def check5(txt):
    for i in txt:
        first = input("%s: " %i)
        if first[0] != i:
            return False
    return True
```

### 승식이의 전동킥보드는 세계제일?


<center>사진설명 : 출근하는 윤승식 조교, 여러분은 킥보드 탈때 핸드폰 쓰지 마세요!</center>

윤승식 조교의 잇템! 전동킥보드를 완충하는 시간은 6시간입니다.

승식이의 킥보드는 완충 시, 30km를 갈 수 있습니다. 

승식이가 가야하는 거리 A와 왕복 여부를 입력받고 이에 필요한 충전 시간을 계산하는 함수 kickboard()를 만들어봅시다.

**기본값은 "편도" 입니다.**

이때, 최대거리 이상 입력할 경우, "킥보드로 갈 수 없습니다" 라는 메시지를 출력하고, return은 None으로 합니다.

```python
def kickboard():
    dist = input("가야하는 거리를 입력하시오(km): ")
    wang = input("왕복여부를 입력하시오(Y/N): ")
    if wang.upper() == "Y":
        wang=2
    else:
        wang=1
        
    if float(dist)*wang > 30:
        print("킥보드로 갈 수 없습니다.")
        return None
    else:
        return float(dist)*wang*6/30
```
