```python
#필요한 패키지를 import
import matplotlib.pyplot as plt
import seaborn as sns

#jupyter에서 matplotlib을 사용해 만든 graph를 화면에 표시하기 위해 필요
%matplotlib inline 

#아래는 한글을 사용할 때 깨지는 문제에 대한 해결
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

#그래프의 축 등에서 음수를 표시할 때 minus sign이 깨지는 것 해결
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
```

```python
tit = sns.load_dataset("titanic")
tit.head(5)
```

```python
sns.countplot(x="class", data=tit)
```

```python
sns.countplot(x="sex", data=tit)
```

```python
sns.barplot(x="sex", y="survived", data=tit)
```

```python
sns.barplot(x="class", y="survived", data=tit)
```

```python
# 연속형 변수에 대한 그래프 그리기
plt.figure(figsize=(8,3))
sns.histplot(x="age", bins=20, kde=True, data=tit)
```

```python
plt.figure(figsize=(8,3))
sns.histplot(x="fare", bins=50, kde=True, data=tit)
```

```python
plt.figure(figsize=(8,3))
sns.histplot(x="age", bins=20, kde=True, hue="survived", data=tit)
```

```python
plt.figure(figsize=(8,3))
sns.histplot(x="age", bins=20, kde=True, stat="percent", data=tit)
```

```python
plt.figure(figsize=(8,3))
sns.histplot(x="age", bins=20, kde=True, hue="survived", multiple="dodge", data=tit)
```

```python
plt.figure(figsize=(8,3))
sns.histplot(x="age", bins=20, kde=True, hue="sex", multiple="dodge", data=tit)
```

```python
#class 별로 살아남은 비중에 대해 막대그래프로 그리기

sns.barplot(x="class", y="survived", hue="sex", data=tit)
```

```python
sns.boxplot(x="class", y="age", hue="sex", data=tit)
```

```python
tips = sns.load_dataset("tips")
tips.head(5)
```

```python
sns.barplot(x="day", y="total_bill", hue="smoker", data=tips)
```

```python
sns.barplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set2")
```

```python
sns.pairplot(hue="smoker", data=tips)
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris)
```

```python
corr=tit.corr()
sns.heatmap(corr, annot=True, fmt=".1f")
corr
```

```python
flights = sns.load_dataset("flights")
print(flights.head(5))
flights = flights.pivot("month", "year", "passengers")
plt.figure(figsize=(10, 10))
ax = sns.heatmap(flights, annot=True, fmt="d")
print(flights.head(5))
```
